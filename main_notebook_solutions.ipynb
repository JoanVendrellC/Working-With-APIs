```python
def flatten(data):
    for column in data:
        flattened = pd.DataFrame(dict(data[column])).transpose()
        flattened.columns = [column + '_' + str(colname) for colname in flattened.columns]
        data = pd.concat([data, flattened], axis=1)
        data = data.drop(column, axis=1)
    return data
```
